### Scatterplot

```python
plt.figure(figsize=(12,10))

plt.subplot(4,2,1)
plt.scatter(idata['SL'],idata['PL'],s=10 ,c ='red', marker='^')
plt.title('Scatterplot SL Vs PL', fontsize = 15)
plt.xlabel('SL', fontsize = 10)
plt.ylabel('PL', fontsize = 10)



plt.subplot(4,2,2)
plt.scatter(idata['SW'],idata['PW'],s=10,c='purple')
plt.title('Scatterplot SW Vs PW', fontsize = 15)
plt.xlabel('SW', fontsize = 10)
plt.ylabel('PW', fontsize = 10)

plt.subplot(4,2,3)
plt.scatter(idata['SL'],idata['PW'],s=10,c='green',marker='*')
plt.title('Scatterplot SL Vs PW', fontsize = 15)
plt.xlabel('SL', fontsize = 10)
plt.ylabel('PW', fontsize = 10)

plt.subplot(4,2,4)
plt.scatter(idata['SW'],idata['PL'],s=10,c='brown',marker='^')
plt.title('Scatterplot SW Vs PL', fontsize = 15)
plt.xlabel('SW', fontsize = 10)
plt.ylabel('PL', fontsize = 10)

plt.tight_layout()
```

```python
#Scatterplot
colors =['red', 'orange','blue']
classification = ['Iris-virginica','Iris-versicolor','Iris-setosa']
```

```python
for i in range(3):
    x=idata[idata['Classification']==classification[i]]
    plt.scatter(x['SL'],x['SW'], c=colors[i], label=classification[i])
plt.xlabel('SL')
plt.ylabel('SW')
plt.legend()
```

```python
for i in range(3):
    x=idata[idata['Classification']==classification[i]]
    plt.scatter(x['PL'],x['PW'], c=colors[i], label=classification[i])
plt.xlabel('PL')
plt.ylabel('PW')
plt.legend()
```
