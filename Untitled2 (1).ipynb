# Case study 1- Presentation of Data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Read the dataset to the python environment.

```python
cars_data= pd.read_csv('cars_data.csv')
cars_data
```

2. Check for the null values present in the dataset.

```python
cars_data.isnull().sum()
```

3. Plot a bar graph of male vs female buyers participated in the sales.

```python
cars_data['Buyer Gender'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=cars_data,x='Buyer Gender')
plt.title('Participation of Male Vs Female',fontweight='bold',fontsize=16)
plt.xlabel('Gender',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=10)
plt.grid()
plt.show()
```

4. Find the top 5 cars based on their sales price.

```python
cars_data.nlargest(5,['Sale Price'])['Make']
```

```python
#corresponding sale Price
sp=cars_data['Sale Price'].sort_values(ascending=False)
sp.head()
```
